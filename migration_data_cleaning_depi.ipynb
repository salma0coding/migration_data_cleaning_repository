```python
# begin cleaning
import pandas as pd
import numpy as np
import re
```

```python
def split_multi_value_cell(value):
    if pd.isna(value):
        return []

    value = str(value).strip()

    value = re.sub(r"\s*/\s*", ",", value)
    value = re.sub(r"\s*;\s*", ",", value)
    value = re.sub(r"\s+\band\b\s+", ",", value, flags=re.IGNORECASE)
    value = re.sub(r"\s*&\s*", ",", value)

    parts = [x.strip() for x in value.split(",")]
    parts = [x for x in parts if x != ""]

    return parts
```

```python
# =========================
# SAFE SPLIT REGION + COUNTRY TOGETHER
# =========================

clean_df["Region_List"] = clean_df["Region of Origin"].apply(split_multi_value_cell)
clean_df["Country_List"] = clean_df["Country of Origin"].apply(split_multi_value_cell)

clean_df["Region_Count"] = clean_df["Region_List"].apply(len)
clean_df["Country_Count"] = clean_df["Country_List"].apply(len)

clean_df["Pairing_Status"] = np.where(
    clean_df["Region_Count"] == clean_df["Country_Count"],
    "Matched",
    "Needs_Review"
)

matched_rows = clean_df[clean_df["Pairing_Status"] == "Matched"].copy()
review_rows = clean_df[clean_df["Pairing_Status"] == "Needs_Review"].copy()

expanded_rows = []

for _, row in matched_rows.iterrows():
    regions = row["Region_List"]
    countries = row["Country_List"]

    if len(regions) == 1 and len(countries) == 1:
        new_row = row.copy()
        new_row["Region of Origin"] = regions[0]
        new_row["Country of Origin"] = countries[0]
        expanded_rows.append(new_row)
    else:
        for r, c in zip(regions, countries):
            new_row = row.copy()
            new_row["Region of Origin"] = r
            new_row["Country of Origin"] = c
            expanded_rows.append(new_row)

matched_expanded_df = pd.DataFrame(expanded_rows)

clean_df = pd.concat([matched_expanded_df, review_rows], ignore_index=True)

# clean again
clean_df["Region of Origin"] = clean_df["Region of Origin"].astype(str).str.strip()
clean_df["Country of Origin"] = clean_df["Country of Origin"].astype(str).str.strip()
```

```python
# Print cleaned column names
print("cleaned columns' names")
print(clean_df.columns.tolist())
```

```python
# strip leading/trailing spaces
#  replace repeated spaces with a single space
text_cols  = clean_df.select_dtypes(include = "object").columns
for col in text_cols:
    clean_df[col] = clean_df[col].apply(
        lambda x: x.strip() if isinstance(x, str) else x
    )
    clean_df[col] = clean_df[col].replace(r"\s+", " ", regex=True)
```

```python
# Define columns that must exist
required_columns = ["Incident Year", "Region of Origin"]

for col in required_columns:
    if col not in clean_df.columns:
        raise ValueError(f"Required column not found: {col}")
```

```python
# Create a cleaned version of Region of Origin:
# lowercase + stripped spaces
clean_df["Region of Origin Clean"] = clean_df["Region of Origin"].astype(str).str.strip().str.lower()
clean_df["Region of Origin Clean"] = clean_df["Region of Origin Clean"].str.replace(r"\s*\(p\)", "", regex=True)
```

```python
# Define African region keywords to keep
africa_keywords = [
    "eastern africa",
    "western africa",
    "northern africa",
    "southern africa",
    "middle africa",
    "sub-saharan africa"
]
# Build one regex pattern from the keywords
africa_pattern = "|".join(africa_keywords)
```

```python
# Keep only rows whose Region of Origin matches African regions
clean_df = clean_df[clean_df["Region of Origin Clean"].str.contains(africa_pattern, na=False)].copy()
```

```python
# Function to assign numeric region code based on region text
def assign_africa_region_code(region_text):
    if pd.isna(region_text):
        return np.nan

    region_text = str(region_text).lower().strip()

    if "eastern africa" in region_text:
        return 3
    elif "western africa" in region_text:
        return 4
    elif "northern africa" in region_text:
        return 1
    elif "southern africa" in region_text:
        return 6
    elif "middle africa" in region_text:
        return 5
    elif "sub-saharan africa" in region_text:
        return 2
    else:
        return np.nan
# Apply the function to generate numeric region codes
clean_df["Africa_Region_Code"] = clean_df["Region of Origin Clean"].apply(assign_africa_region_code)
```

```python
region_label_map = {
    3: "East Africa",
    4: "West Africa",
    1: "North Africa",
    6: "South Africa",
    5: "Central Africa",
    2: "Sub-Saharan Africa"
}

clean_df["Africa_Region_Name"] = clean_df["Africa_Region_Code"].map(region_label_map)
```

```python
import pandas as pd
import numpy as np

# =========================
# LOAD FILE
# =========================
file_name = "cleaned_africa_data (3).xlsx"   # change if needed
df = pd.read_excel(file_name)

# =========================
# CHECK REQUIRED COLUMNS
# =========================
required_cols = [
    "Numbers_Found_Count",
    "Number of Females",
    "Number of Males",
    "Number of Children"
]

for col in required_cols:
    if col not in df.columns:
        raise ValueError(f"Required column not found: {col}")

# =========================
# REPLACE / RENAME SURVIVOR COLUMN
# =========================
# If old 'Number of Survivors' exists, remove it first
if "Number of Survivors" in df.columns:
    df.drop(columns=["Number of Survivors"], inplace=True)

# Rename Numbers_Found_Count -> Number of Survivors
df.rename(columns={"Numbers_Found_Count": "Number of Survivors"}, inplace=True)

# Make sure numeric
df["Number of Survivors"] = pd.to_numeric(df["Number of Survivors"], errors="coerce").fillna(0).astype(int)
df["Number of Females"] = pd.to_numeric(df["Number of Females"], errors="coerce").fillna(0)
df["Number of Males"] = pd.to_numeric(df["Number of Males"], errors="coerce").fillna(0)
df["Number of Children"] = pd.to_numeric(df["Number of Children"], errors="coerce").fillna(0)

# =========================
# FUNCTION TO REDISTRIBUTE
# =========================
def redistribute_survivors(row):
    total_survivors = int(row["Number of Survivors"])

    f = float(row["Number of Females"])
    m = float(row["Number of Males"])
    c = float(row["Number of Children"])

    original_sum = f + m + c

    # Case 1: no survivors
    if total_survivors == 0:
        return pd.Series([0, 0, 0])

    # Case 2: no old sex/age info -> equal split
    if original_sum == 0:
        base = total_survivors // 3
        remainder = total_survivors % 3

        females_new = base
        males_new = base
        children_new = base

        # distribute remainder
        if remainder >= 1:
            females_new += 1
        if remainder >= 2:
            males_new += 1

        return pd.Series([females_new, males_new, children_new])

    # Case 3: use original ratio
    raw_f = total_survivors * (f / original_sum)
    raw_m = total_survivors * (m / original_sum)
    raw_c = total_survivors * (c / original_sum)

    # floor first
    females_new = int(np.floor(raw_f))
    males_new = int(np.floor(raw_m))
    children_new = int(np.floor(raw_c))

    assigned = females_new + males_new + children_new
    remainder = total_survivors - assigned

    # fractional parts for fair remainder distribution
    fractions = [
        ("Number of Females", raw_f - females_new),
        ("Number of Males", raw_m - males_new),
        ("Number of Children", raw_c - children_new),
    ]

    fractions.sort(key=lambda x: x[1], reverse=True)

    for i in range(remainder):
        if fractions[i][0] == "Number of Females":
            females_new += 1
        elif fractions[i][0] == "Number of Males":
            males_new += 1
        else:
            children_new += 1

    return pd.Series([females_new, males_new, children_new])

# =========================
# APPLY REDISTRIBUTION
# =========================
df[["Number of Females", "Number of Males", "Number of Children"]] = df.apply(
    redistribute_survivors, axis=1
)

# =========================
# OPTIONAL CHECK COLUMN
# =========================
df["Check_Total"] = (
    df["Number of Females"] +
    df["Number of Males"] +
    df["Number of Children"]
)

# =========================
# SAVE
# =========================
output_file = "cleaned_africa_data_updated.xlsx"
df.to_excel(output_file, index=False)

print("Done.")
print("Saved as:", output_file)
```
